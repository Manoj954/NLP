```python
import time,os
from datetime import datetime
import pandas as pd
import json
import ast

path="D:\\Projects\\Data\\News\\Data\\"
category=["business","entertainment","general","health","science","sports","technology"]
columns=["description","content","author","title","category"]
all_files=[]
for files in os.listdir(path):
    if "2019" in files:
        all_files.append(os.path.join(path,files))

df_new=[]
for files in all_files:
    print(files)
    df=pd.read_csv(files,header=None,delimiter='~')
    for item in df[0]:
        #print(item)
        obj=ast.literal_eval(item)
        df_new.append([obj["description"],obj["content"],obj["author"],obj["title"],str(files).split('\\')[-1].split('_')[-1].replace('.txt','')])
        

df_new=pd.DataFrame(df_new,columns=columns)
df_new.to_csv(path+"\\Top_Headlines.csv",sep='~')
```
