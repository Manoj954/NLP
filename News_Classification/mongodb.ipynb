```python
def fetchData(stDate,endDate):
    text = db.datetime.aggregate([{"$match":{"ts":{"$elemMatch":{"$gte":stDate,"$lte":endDate}}}},
                {"$unwind":"$ts"},
                {"$match":{"ts":{"$gte":stDate,"$lte":endDate}}},
                {"$group":{"_id":"$_id","count":{"$sum":1}}}])
    jsondata = text["result"]
    for each in jsondata:
        try:
            print(each)  
        except Exception:
            print('err')   
```

```python
fetchData("2019:04:06 05:36:10","2019:04:06 04:36:10")
```
