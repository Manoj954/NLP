```python
import os,time
import pandas as pd
from datetime import datetime 
import pickle

#preprocessing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

#Model Selection
from sklearn.model_selection import train_test_split

#ML models
# Multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB
#NN
from sklearn.neural_network import MLPClassifier
#SVM
from sklearn import svm

#Evalution Metrics
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
#Load Data
path="D:\\Projects\\Data\\News\\Data\\"
df=pd.read_csv(os.path.join(path,"Top_Headlines.csv"),sep='~', encoding='utf-8')
```

```python
#EDA
df.head()
```

```python
df=df.drop(df.columns[0],axis=1)
```

```python
#df.description=str(df.description)
#df=df.apply(lambda x: str(x))
```

```python
df.head()
```

```python
#print(df.description.unique)
```

```python
#GET VECTOR COUNT
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(df.description.values.astype('U'))

#SAVE WORD VECTOR
#pickle.dump(count_vect.vocabulary_, open("count_vector.pkl","wb"))
```

```python

#TRANSFORM WORD VECTOR TO TF IDF
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

#SAVE TF-IDF
#pickle.dump(tfidf_transformer, open("tfidf.pkl","wb"))
```

```python


#clf = MultinomialNB().fit(X_train_tfidf, training_data.flag)
X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, df.category, test_size=0.25, random_state=42)
clf = MultinomialNB().fit(X_train, y_train)

#SAVE MODEL
#pickle.dump(clf, open("nb_model.pkl", "wb"))
```

```python
#test
```

```python
predicted = clf.predict(X_test)
result_bayes = pd.DataFrame( {'true_labels': y_test,'predicted_labels': predicted})
#result_bayes.to_csv('res_bayes.csv', sep = ',')

#for predicted_item, result in zip(predicted, y_test):
#    print(category_list[predicted_item], ' - ', category_list[result])
```

```python
confusion_mat = confusion_matrix(y_test,predicted)
print(confusion_mat)
```

```python

clf_neural = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15,), random_state=1)

#X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, training_data.flag, test_size=0.25, random_state=42)

clf_neural.fit(X_train, y_train)
```

```python
predictedNN = clf_neural.predict(X_test)
result_softmax = pd.DataFrame( {'true_labels': y_test,'predicted_labels': predictedNN})
#result_softmax.to_csv('res_softmax.csv', sep = ',')

#for predicted_item, result in zip(predicted, y_test):
#    print(category_list[predicted_item], ' - ', category_list[result])
```

```python
clf_svm = svm.LinearSVC()
#X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, training_data.flag, test_size=0.25, random_state=42)
clf_svm.fit(X_train_tfidf, df.category)
#pickle.dump(clf_svm, open("svm.pkl", "wb"))
```

```python
predictedsvm = clf_svm.predict(X_test)
result_svm = pd.DataFrame( {'true_labels': y_test,'predicted_labels': predictedsvm})
#result_svm.to_csv('res_svm.csv', sep = ',')
#for predicted_item, result in zip(predicted, y_test):
#    print(category_list[predicted_item], ' - ', category_list[result])
```

```python
print(classification_report(y_test, predicted))
print(classification_report(y_test, predictedNN))
print(classification_report(y_test, predictedsvm))
```
